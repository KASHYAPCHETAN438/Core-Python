# **What is File Handling?**


**Definition:**
  -  File handling is the process of creating, opening, reading, writing, and closing files through a program.
  - The purpose of Files concept is that "To store the data Permanently".
  - The process of storing the data permanently is knownas "Persistency".


**Data Persistence**

- Persistence is the ability to store data as an object in a permanent location such as a File or a Database.
- File is used to save the data permanently on a disk.
- File is a collection of data stored on a disk with a specific name and directory
path.

**Why We Need File Handling:**

- Variables store data temporarily (in RAM).

- Once the program ends, that data is lost.

- File handling allows us to store data permanently on disk so we can use it later.


# **Applications Based on File Handling**

1. **Non-Persistent Applications**

    - **Definition:** Applications that store data temporarily in memory.

    - **Flow:**

        - Accept data from the user (e.g., via keyboard).

        - Store data in main memory (RAM) as objects/variables.

        - Process the data and display results on the console.

    - **Key Point:**

       - Data is temporary — lost when the program ends.

    - **Example:**

        - A calculator program that runs and shows result, but doesn’t save history.


2. **Persistent Applications**

    - **Definition:** Applications that store data permanently so it can be reused later.

    - **Flow:**

       - Accept data from the user.

       - Store data in memory as objects temporarily.

       - Process the data.

        - Save the results permanently in files or databases.

    - **Key Point:**

        - Data is persistent — survives even after the program ends.

# **Ways to Store Data Permanently**

1. **Using Files**

   - Store data in text or binary files on disk.

   - Simple to use for small/medium programs.

   - Example: .txt, .csv, .json, .dat.

2. **Using Database Software**

    - Store structured data in databases like MySQL, SQLite, MongoDB.

    - Used for large-scale applications requiring complex queries and relationships.

# **We can open file by two ways**

1. **File Opening using open()**

   - open() is one of the pre-defined functions present in the default module builtins.

    - This function is used for opening any specified file in a specified file mode.

   - Syntax:

          varname = open("FileName", "FileMode")


    - Explanation:

        - varname is an object of TextIOWrapper and it points to the file; it is called a File Pointer.

        - File Name represents the name of the file.

       - File Mode can be any of the following:

        - r, w, a, r+, w+, a+, x

   - Once we open the file with open(), it is mandatory to close the file using close(); this is a manual process.

   - open() does not support **automatic closing** of files.






```python
filename = input("Enter filename (With Path): ")

try:
    fp = open(filename, "r")
    content = fp.read()
    print(content)
    fp.close()                          # Manual closing
except FileNotFoundError:
    print("File does not exist")
```

2. **File Opening using with open()**

   - with and as are keywords in Python.

   - open() is a pre-defined function used to open a file in a specified mode.

   - After the `with` block ends, the file is **automatically closed**.

  - Syntax:

        with open("FileName", "FileMode") as varname:
        # Block of statements – File operations


    - **varname** → File pointer (TextIOWrapper) pointing to the file.

    - Block of statements → Set of executable statements for performing file operations.

    - **Advantages of with open() as:**

      - File remains open inside the block.

      - File is automatically closed when execution exits the block.

     - No need for finally block or manual close().

     - Makes code cleaner and safer.

**Attribute and Method**


1. **type(fp)** → <class '_io.TextIOWrapper'>
    - _io.TextIOWrapper is a Python class that handles **text files**.  
    - Provides file operations like read(), write(), seek(), etc.  

2. fp.name
    - Returns the **actual path or name** of the file.  

3. fp.readable()
    - Returns **True** if the file is opened in **read mode**.  

4. fp.writable()
    - Returns **True** if the file is opened in **write mode**, otherwise **False**.  

5. fp.closed
    - Inside the with block → **False** (file is still open)  
    - Outside the with block → **True** (file is automatically closed)  




```python
# Open a file and read its content using 'with open()'
with open("PythonFileHandling.txt", "r") as fp:          # fp is the file pointer
    content = fp.read()
    print(content)

# No need to manually close the file
print("Is file closed ?", fp.closed)
```

# File Opening Modes in Python

## 1) `r` mode
- Opens the file in **READ mode**.  
- This is one of the **default modes**.  
- If the file does not exist, raises a **FileNotFoundError**.
- **Cursor Position:** At the **beginning** of the file


```python
try:
    with open("PythonFileHandling.txt") as fp:                           # by default 'r' mode
        print("----- File Created and Opened in Read Mode -----\n")
        print("Type of the file:", type(fp))                             # TextIOWrapper ke bare me likh
        print("File Name:", fp.name)
        print("Current Cursor Position : ",fp.tell())                   # Don't get confused if file empty cursur position is 0
        print("File Readable:", fp.readable())
        print("File Writable:", fp.writable())
        print("File Closed (inside with):", fp.closed)                 # False inside with block

    # After the with block
    print("File Closed (outside with):", fp.closed)                    # True after exiting with block

except FileNotFoundError:
    print("File not found")
```


---

## 2) `w` mode
- Used for **creating a new file** and always opens in **WRITE mode**.  
- If the file already exists, its content is **overwritten**.  
- **Cursor Position:** At the **beginning** of the file (existing data is erased).



```python
with open("PythonFileHandling.txt", "w") as fp:  # Write mode
    print("\n----- w mode: Write Mode -----\n")
    print("Type:", type(fp))
    print("File Name:", fp.name)
    print("Current Cursor Position : ",fp.tell())
    print("Readable:", fp.readable())
    print("Writable:", fp.writable())
    print("Closed (inside):", fp.closed)
print("Closed (outside):", fp.closed)
```

---

## 3) `a` mode
- Used for opening a new file and performing **WRITE operations**.  
- Can also open an existing file in **WRITE mode**, and new data will be **appended**.  
 **Cursor Position:** At the **end** of the file (new data is appended).



```python
with open("PythonFileHandling.txt", "a") as fp:  # Append mode
    print("\n----- a mode: Append Mode -----")
    print("Type:", type(fp))
    print("File Name:", fp.name)
    print("Current Cursor Position : ",fp.tell())
    print("Readable:", fp.readable())
    print("Writable:", fp.writable())
    print("Closed (inside):", fp.closed)
print("Closed (outside):", fp.closed)
```

---

## 4) `r+` mode
- Opens the file in **READ mode** (if the file exists).
- Allows both **READ and WRITE operations**.   
- **READ happens first**, then WRITE can be performed.  
- **Cursor Position:** At the **beginning** of the file (read/write starts from start).

---


```python
try:
    with open("PythonFileHandling.txt", "r+") as fp:  # Read and write mode
        print("\n----- r+ mode: Read and Write -----")
        print("Type:", type(fp))
        print("File Name:", fp.name)
        print("Current Cursor Position : ",fp.tell())
        print("Readable:", fp.readable())
        print("Writable:", fp.writable())
        print("Closed (inside):", fp.closed)
    print("Closed (outside):", fp.closed)
except FileNotFoundError:
    print("File not found for r+ mode")
```


## 5) `w+` mode
- Opens the file in **WRITE mode**.  
- First **writes**, then allows **READ operations**.  
- If the file exists, its content is **overwritten**.  
- **Cursor Position:** At the **beginning** of the file.


```python
with open("PythonFileHandling.txt", "w+") as fp:  # Write and read mode
    print("\n----- w+ mode: Write and Read -----\n")
    print("Type:", type(fp))
    print("File Name:", fp.name)
    print("Current Cursor Position : ",fp.tell())
    print("Readable:", fp.readable())
    print("Writable:", fp.writable())
    print("Closed (inside):", fp.closed)
print("Closed (outside):", fp.closed)
```


---

## 6) `a+` mode
- Used for opening a new file in **WRITE mode**.  
- Can also open an existing file, and new data will be **appended**.  
- Additionally allows **READ operations** (first WRITE/append, then READ).  
- **Cursor Position:** At the **end** of the file (for writing).  
- **Note:** For reading, you need to manually move the cursor (e.g., using seek(0)).


```python
with open("PythonFileHandling.txt", "a+") as fp:  # Append and read mode
    print("\n----- a+ mode: Append and Read -----\n")
    print("Type:", type(fp))
    print("File Name:", fp.name)
    print("Current Cursor Position : ",fp.tell())
    print("Readable:", fp.readable())
    print("Writable:", fp.writable())
    print("Closed (inside):", fp.closed)
print("Closed (outside):", fp.closed)
```


---

## 7) `x` mode
- Exclusively used for **creating a new file in WRITE mode**.  
- The file can be opened **only once** in this mode.  
- If the file already exists, raises a **FileExistsError**.  
- **Cursor Position:** At the **beginning** of the new empty file.
- Ex : Biometric Thumb attendance (Only write one time a day)


```python
try:
    with open("PythonFileHand.txt", "x") as fp:               # Exclusive create mode
        print("\n----- x mode: Exclusive Create -----\n")
        print("Type:", type(fp))
        print("File Name:", fp.name)
        print("Current Cursor Position : ",fp.tell())
        print("Readable:", fp.readable())
        print("Writable:", fp.writable())
        print("Closed (inside):", fp.closed)
    print("Closed (outside):", fp.closed)
except FileExistsError:
    print("File already exists for x mode")
```


---
## 8) `x+` mode
- Used for **creating a new file in both READ and WRITE mode**.  
- If the file already exists, raises a **FileExistsError**.  
-**Cursor Position:** At the **beginning** of the new empty file.

```python
try:
    with open("FileHandling.txt", "x+") as fp:  # Exclusive create + read/write
        print("\n----- x+ mode: Exclusive Create and Read/Write -----\n")
        print("Type:", type(fp))
        print("File Name:", fp.name)
        print("Current Cursor Position : ",fp.tell())
        print("Readable:", fp.readable())
        print("Writable:", fp.writable())
        print("Closed (inside):", fp.closed)
    print("Closed (outside):", fp.closed)
except FileExistsError:
    print("File already exists for x+ mode")
```

```python
# --------- Text File Modes Example --------------

file_path = "/content/PythonFileHandling.txt"


# r mode → Read only (file must exist)
try:
    with open(file_path, "r") as f:
        print("r mode →", f.read(50))   # read first 50 characters
except FileNotFoundError:
    print("File not found for r mode")



# w mode → Write only (creates file if not exist, truncates if exists)
with open(file_path, "w") as f:
    f.write("Hello from w mode\n")
    print("w mode → File truncated and written")



# a mode → Append only (creates file if not exist, adds data at end)
with open(file_path, "a") as f:
    f.write("Hello from a mode\n")
    print("a mode → Data appended")



# r+ mode → Read + Write (file must exist, does not truncate)
with open(file_path, "r+") as f:
    data = f.read(20)     # read first 20 characters
    print("r+ mode → First 20 chars:", data)
    f.write("\nAdded from r+ mode")



# w+ mode → Write + Read (creates file if not exist, truncates if exists)
with open(file_path, "w+") as f:
    f.write("Written in w+ mode\n")
    f.seek(0)   # go back to start
    print("w+ mode →", f.read())



# a+ mode → Append + Read (creates file if not exist, appends at end)
with open(file_path, "a+") as f:
    f.write("Line added in a+ mode\n")
    f.seek(0)   # go back to start to read entire content
    print("a+ mode →", f.read())



# x mode → Create new (error if file already exists)
try:
    with open("/content/newfile.txt", "x") as f:
        f.write("Created with x mode\n")
        print("x mode → New file created")
except FileExistsError:
    print("x mode → File already exists")
```

```python
# ------------- Binary File Modes Example -----------------


binary_path = "/content/PYTHON-NOTES.pdf"


# rb mode → Read binary (file must exist)
try:
    with open(binary_path, "rb") as f:
        print("rb mode → First 20 bytes:", f.read(20))
except FileNotFoundError:
    print("File not found for rb mode")



# wb mode → Write binary (creates new file or truncates existing)
with open("/content/test_binary.pdf", "wb") as f:
    f.write(b"%PDF-1.4\n%Binary file created\n")
    print("wb mode → Wrote binary data to test_binary.pdf")



# ab mode → Append binary (creates file if not exist, appends at end)
with open("/content/test_binary.pdf", "ab") as f:
    f.write(b"Appended in binary mode\n")
    print("ab mode → Appended binary data")



# rb+ mode → Read + Write binary (file must exist, no truncate)
with open("/content/test_binary.pdf", "rb+") as f:
    content = f.read(10)   # read first 10 bytes
    print("rb+ mode → First 10 bytes:", content)
    f.write(b"\nExtra binary data")



# wb+ mode → Write + Read binary (creates file if not exist, truncates if exists)
with open("/content/test_binary2.pdf", "wb+") as f:
    f.write(b"New binary content")
    f.seek(0)
    print("wb+ mode →", f.read())



# ab+ mode → Append + Read binary (creates file if not exist, append at end)
with open("/content/test_binary2.pdf", "ab+") as f:
    f.write(b"\nAdded with ab+ mode")
    f.seek(0)
    print("ab+ mode →", f.read())



# xb mode → Create new binary (error if file exists)
try:
    with open("/content/unique_binary.pdf", "xb") as f:
        f.write(b"Created new binary file")
        print("xb mode → New binary file created")
except FileExistsError:
    print("xb mode → File already exists")
```

# **Read Operation**

- In Python, the **read operation** allows you to **read data from a file**.  

- A file must be **opened in read mode** (`'r'`, `'r+'`, `'w+'`, `'a+'`) to perform reading.
- File must already exist.

- Does not truncate (clear) the file.

---

## File Read Modes

| Mode  | Description |
|-------|------------|
| `'r'` | Read-only mode. File must exist. |
| `'r+'` | Read and write mode. File must exist. |
| `'w+'` | Write and read mode. Creates file if not exists. |
| `'a+'` | Append and read mode. Creates file if not exists. |

---

## File Object Attributes for Reading

| Attribute | Description |
|-----------|------------|
| `readable()` | Returns `True` if the file can be read. |
| `closed` | Returns `True` if the file is closed. |
| `name` | Returns the file name/path. |
| `mode` | Returns the mode in which the file is opened. |

---

## Common File Read Methods

### 1) **read(size=-1)**
- Reads **entire file** if `size` is not provided.  
- Reads **up to `size` characters** if specified.  
- Returns a **string**.  

**Example:**




```python

try:
  fp=open("PythonFileHandling.txt", "r")
  print("\n")
  content = fp.read()

  #print("File Content:\n", content)            # , add one extra space to remove it use +
  print("File Content:\n"+content)
  print("\nCurrent cursor position : ",fp.tell())


  content=fp.read(20)
  print("\nFile Content:"+ content)
  print("Current cursor position : ",fp.tell())

except FileNotFoundError as fe:
  print("File not exits : ",fe.args)
```

### 2) **readlines(hint=-1)**

- Reads all lines from the file and returns them as a list of strings.

- Optional hint specifies approximate number of bytes to read.

- Each element in the list represents one line, including newline characters.

**Example:**

```python
with open("example.txt", "r") as fp:
    lines = fp.readlines()
    print(lines)
```

- This is memory-efficient for large files.

##  Cursor Methods for Read Operation

### 4) **seek(offset, whence=0)**

- Moves the file cursor to a specific position.

- **Parameters:**

  - offset → Number of bytes/characters to move the cursor.

  - whence → Reference position:

  - 0 → Start of file (default)

  - 1 → Current cursor position

  - 2 → End of file

- Usage: Allows reading from a specific location in the file.

**Example:**


```python
with open("example.txt", "r") as fp:
    fp.seek(5)           # Move cursor to 5th byte from start
    print(fp.read(10))   # Read 10 characters from position 5
```

### 5) **Tell()**

- Returns the current position of the file cursor.

- Useful to know where you are in the file during read operations.

**Example**

```python
with open("example.txt", "r") as fp:
    print("Initial Cursor:", fp.tell())  # 0
    fp.read(5)
    print("Cursor after reading 5 chars:", fp.tell())  # 5
```

# **Write Operation**


- In Python, **write operations** allow you to **write data to a file**.  
- A file must be opened in a **write-compatible mode** (`'w'`, `'w+'`, `'a'`, `'a+'`, `'r+'`, `'x'`, `'x+'`) to perform writing.

- File is created if not exists.

- If file exists, it is truncated (cleared) before writing.

- Opens the file for both writing & reading.

- Writing always starts from the beginning (but file is already empty).

---

## File Write Modes

| Mode  | Description |
|-------|------------|
| `'w'` | Write-only mode. Creates a new file or **overwrites** existing file. |
| `'w+'` | Write and read mode. Creates file if not exists. |
| `'a'` | Append mode. Adds new data at the **end** of the file. |
| `'a+'` | Append and read mode. Creates file if not exists. |
| `'r+'` | Read and write mode. File must exist. |
| `'x'` | Exclusive creation mode. Fails if file already exists. |
| `'x+'` | Exclusive creation + read/write. Fails if file exists. |

---

## File Object Attributes for Writing

| Attribute | Description |
|-----------|------------|
| `writable()` | Returns `True` if the file can be written. |
| `closed` | Returns `True` if the file is closed. |
| `name` | Returns the file name/path. |
| `mode` | Returns the mode in which the file is opened. |

---

## Common File Write Methods

### 1) **write(string)**
- Writes the given **string** to the file.  
- Does **not add a newline character** automatically.  
- Returns the **number of characters written**.  

**Example:**


```python

with open("example_write.txt", "w") as fp:
    count = fp.write("Hello World!\n")
    print("Characters Written:", count)
```


### 2) **writelines(list_of_strings)**

- Writes a list of strings to the file.

- Does not add newlines automatically, so strings should include \n if needed.

**Example**

```python
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
with open("example_write.txt", "w") as fp:
    fp.writelines(lines)
```


### 3) **Iterative Writing Using a for loop**

- You can iterate over a list of strings or any iterable and write each line.

**Example**

```python
lines = ["Python\n", "File Handling\n", "Write Operation\n"]
with open("example_write.txt", "w") as fp:
    for line in lines:
        fp.write(line)
```

```python
# Seek()

with open("example_write.txt", "w+") as fp:
    fp.write("Hello World")
    fp.seek(6)            # Move cursor to overwrite from position 6
    fp.write("Python")
```

```python
# Tell()

with open("example_write.txt", "w") as fp:
    fp.write("Hello")
    print("Cursor position:", fp.tell())  # 5
```
