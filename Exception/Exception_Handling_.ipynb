# **-------Exception & Error Handling---------**

# Error
  * Mistake in program code (syntax or logic)
  * Compile time (before program runs).
  * Cannot be handled with try-except.
  * program stops immediately.
  * **Cause**
        
       * Wrong syntax, indentation, invalid structure.

  * **Example**
    * SyntaxError, IndentationError, NameError (undeclared variable).

  * SyntaxError

  * IndentationError

  * TabError


# Exception
  * Issue that happens during runtime execution
  * Runtime (while program is running)
  
  * Can be handled with try-except
  * program can continue after handling
  * **Cause**
        
       * Wrong operations like divide by zero, invalid type conversion, missing key.

  * **Example**
    * ZeroDivisionError, ValueError, IndexError, FileNotFoundError






# Types of Exception

image.png

```python

"""
Examples of different Python Exceptions
as per the hierarchy diagram
"""

# 1. ArithmeticError (Base class)
try:
    raise ArithmeticError("Arithmetic error occurred")
except ArithmeticError as e:
    print("ArithmeticError Example:", e)



# 2. ZeroDivisionError (Subclass of ArithmeticError)
try:
    x = 10 / 0
except ZeroDivisionError as e:
    print("\nZeroDivisionError Example:", e)



# 3. FloatingPointError (Subclass of ArithmeticError)
import math
try:
    raise FloatingPointError("Invalid floating point operation")
except FloatingPointError as e:
    print("\nFloatingPointError Example:", e)



# 4. NameError
try:
    print(undeclared_var)
except NameError as e:
    print("\nNameError Example:", e)



# 5. ValueError
try:
    int("abc")
except ValueError as e:
    print("\nValueError Example:", e)

```

```python
# 6. ExceptionGroup (Python 3.11+)
try:
    raise ExceptionGroup("Multiple exceptions occurred", [ValueError("val error"), TypeError("type error")])
except ExceptionGroup as e:
    print("\nExceptionGroup Example:", e)

# 7. TypeError
try:
    result = "2" + 2
except TypeError as e:
    print("\nTypeError Example:", e)

# 8. OSError (Base class for OS-related errors)
try:
    raise OSError("OS related error")
except OSError as e:
    print("\nOSError Example:", e)

# 9. PermissionError (Subclass of OSError)
try:
    with open("/root/secret.txt", "w") as f:  # restricted in most systems
        f.write("test")
except PermissionError as e:
    print("\nPermissionError Example:", e)

# 10. FileNotFoundError (Subclass of OSError)
try:
    open("non_existing_file.txt")
except FileNotFoundError as e:
    print("\nFileNotFoundError Example:", e)


# 11. RuntimeError
try:
    raise RuntimeError("General runtime error")
except RuntimeError as e:
    print("\nRuntimeError Example:", e)
```


# How to handle Exception

# try  

  * Meaning  

    * Block that contains risky code where an error might occur.  
    * Used to test code execution without crashing the program.  

  * When it is used  

    * When there is a possibility that a code block may raise an error.  

  * Behavior  

    * If an error occurs → control goes to the `except` block.  
    * If no error occurs → the `else` block (if written) executes.  



```python
# Example of try
try:
    print("Before error")
    x = 10 / 0               # risky code
    print("After error")    # not execute
```



# except  

  * Meaning  

    * Block of code that executes when an error occurs in the `try` block.  
    * Used to handle specific or generic exceptions.  

  * When it is used  

    * When an error is raised inside the `try` block.  

  * Behavior  

    * If the raised error matches → this block runs.  
    * If no matching `except` is found → the program stops.  



```python
# Example of except
try:
    x = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero!")
```



# else  

  * Meaning  

    * Block of code that runs only if no error occurs in the `try` block.  

  * When it is used  

    * When you want to execute code only if the `try` block is successful.  

  * Behavior  

    * Executes only when no exception is raised in the `try` block.  



```python
# Example of else
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error: Division by zero")
else:
    print("Division successful:", x)
```



# finally  

  * Meaning  

    * Block of code that always executes regardless of whether an error occurred or not.  

  * When it is used  

    * For cleanup operations like closing files, disconnecting databases, or releasing resources.  

  * Behavior  

    * Always executes, whether an exception was raised or not.  



```python
# Example of finally
try:
    f = open("test.txt", "w")
    f.write("Hello")
except Exception as e:
    print("Error:", e)
finally:
    f.close()
    print("File closed (cleanup done)")
```



# raise  

  * Meaning  

    * Used to manually raise an exception.  

  * When it is used  

    * When you want to throw an error based on a condition in your program.  

  * Behavior  

    * Raises an exception (built-in or custom) and stops normal program flow unless handled.  



```python
# Example of raise
age = 5
if age < 0:
    raise ValueError("Age cannot be negative")
else:
  raise SyntaxError("just for checking")
```

# Difference B/W Exception Handling Keywords

| Keyword   | Meaning / Purpose                         | When it Executes                     | Special Point                                |
|-----------|-------------------------------------------|---------------------------------------|----------------------------------------------|
| try       | Block that contains risky code            | Always executed first                 | Used to test code that may raise an error    |
| except    | Handles the error if it occurs in `try`   | Runs only when an error occurs        | Must match the raised error type             |
| else      | Runs if no error occurs in `try`          | Only when `try` is successful         | Skipped if an exception is raised            |
| finally   | Cleanup code (always runs)                | Runs whether error occurs or not      | Mostly used for resource cleanup (files, DB) |
| raise     | Manually throw an exception               | When a condition in code fails        | Stops program unless handled by try-except   |
